```python
# YOUR CODE HERE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, log_loss
```

```python
X = df_encoded
y = df['income_binary']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.10)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def train_test_LR(X_train, X_test, y_train, y_test, c=1):
    model = LogisticRegression(C=c, max_iter=1000,class_weight='balanced')
    model.fit(X_train, y_train)
    proba_preds = model.predict_proba(X_test)
    loss = log_loss(y_test, proba_preds)
    class_preds = model.predict(X_test)
    acc = accuracy_score(y_test, class_preds)
    return acc, loss

# Hyperparameter range
cs = [10**i for i in range(-5, 5)]

scores = []
losses = []

for c in cs:
    acc, loss = train_test_LR(X_train_scaled, X_test_scaled, y_train, y_test, c)
    scores.append(acc)
    losses.append(loss)
    print('*** value of C:',c)
    print('accuracy score is: ',acc)
    print('loss is: ',loss)
```

```python
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(max_iter=1000,class_weight='balanced')

param_grid = {'C': [10**i for i in range(-10, 10)]}

grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')

grid.fit(X_train_scaled, y_train)

print("Best C found:", grid.best_params_['C'])
print("Best cross-validation accuracy:", grid.best_score_)

# Best model
best_model = grid.best_estimator_
test_preds = best_model.predict(X_test_scaled)
test_acc = accuracy_score(y_test, test_preds)
print("Test set accuracy with best model:", test_acc)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(f_classif, k=13)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

best_model_fs = LogisticRegression(C=grid.best_params_['C'], max_iter=1000,class_weight='balanced')
best_model_fs.fit(X_train_selected, y_train)

preds_fs = best_model_fs.predict(X_test_selected)
acc_fs = accuracy_score(y_test, preds_fs)
print("Test set accuracy after feature selection:", acc_fs)

```
